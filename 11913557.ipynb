```python
import pandas as pd
ageData=pd.read_excel("AgeGroupDetails.xlsx")
ageData
```

```python
import matplotlib.pyplot as pltpie

ageData['TotalCases'].plot(kind='pie',
                            figsize=(15,15),
                            autopct='%1.1f%%', 
                            startangle=70,     
                 
                            
                            )

pltpie.title('Total Beds in Different State and UT')
pltpie.axis('equal') # Sets the pie chart to look like a circle.

pltpie.show()
```

```python
import numpy as np
Age_data = ageData[[ "AgeGroup","TotalCases", "Percentage"]].drop_duplicates()
Age_data.shape
```

```python
boxplot = Age_data.boxplot(['Percentage', 'TotalCases'])
```

```python
import matplotlib.pyplot as pltbar
x=ageData['TotalCases']
y=ageData['Percentage']

pltbar.xlabel('Total Cases')
pltbar.ylabel('percentage')
pltbar.title("Total Corona Cases according to Percentage")
pltbar.bar(x,y)
```
